```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
ds = pd.read_csv('OppScrData.csv')
ds.head()
```

```python
Clinical_Data = ds.iloc[0:9223,3:14]
Clinical_Data.head(10)
```

```python
# Clinical F/U interval [d from CT]
Clinical_Data.loc[ds['Clinical F/U interval  [d from CT]'] < 20, 'Clinical F/U interval  [d from CT]'] = 0
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 30) & ((ds['Clinical F/U interval  [d from CT]']) < 180), 'Clinical F/U interval  [d from CT]'] = 1 # 1-6 month
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 180) & ((ds['Clinical F/U interval  [d from CT]']) < 365), 'Clinical F/U interval  [d from CT]'] = 2 # 0.5-1 years
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 365) & ((ds['Clinical F/U interval  [d from CT]']) < 1095), 'Clinical F/U interval  [d from CT]'] = 3 # 1-3 years
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 1095) & ((ds['Clinical F/U interval  [d from CT]']) < 1825), 'Clinical F/U interval  [d from CT]'] = 4 # 3-5 years
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 1825) & ((ds['Clinical F/U interval  [d from CT]']) < 2555), 'Clinical F/U interval  [d from CT]'] = 5 # 5-7 years
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 2555) & ((ds['Clinical F/U interval  [d from CT]']) < 3650), 'Clinical F/U interval  [d from CT]'] = 6 # 7-10 years
Clinical_Data.loc[((ds['Clinical F/U interval  [d from CT]']) >= 3650) & ((ds['Clinical F/U interval  [d from CT]']) < 4380), 'Clinical F/U interval  [d from CT]'] = 7 # 7-12 years
Clinical_Data.loc[ds['Clinical F/U interval  [d from CT]'] < 5475, 'Clinical F/U interval  [d from CT]'] = 8 # 12-15 years

# BMI
Clinical_Data.loc[ds['BMI'] < 20, 'BMI'] = 0
Clinical_Data.loc[((ds['BMI']) >= 20) & ((ds['BMI']) < 30), 'BMI'] = 1
Clinical_Data.loc[((ds['BMI']) >= 30) & ((ds['BMI']) < 40), 'BMI'] = 2
Clinical_Data.loc[((ds['BMI']) >= 40) & ((ds['BMI']) < 50), 'BMI'] = 3
Clinical_Data.loc[((ds['BMI']) >= 50) & ((ds['BMI']) < 60), 'BMI'] = 4
Clinical_Data.loc[((ds['BMI']) >= 60) & ((ds['BMI']) < 70), 'BMI'] = 5
Clinical_Data.loc[((ds['BMI']) >= 70), 'BMI'] = 6

# BMI >30
Clinical_Data.loc[ds['BMI >30'] == 'N', 'BMI >30'] = 0
Clinical_Data.loc[ds['BMI >30'] == 'Y', 'BMI >30'] = 1
Clinical_Data['BMI >30'] = Clinical_Data['BMI >30'].fillna('-') # -: Missing/not know


# Sex
Clinical_Data.loc[ds['Sex'] == 'Female', 'Sex'] = 0
Clinical_Data.loc[ds['Sex'] == 'Male', 'Sex'] = 1

# Age at CT: Nomal Values(18.5 to 24.9)
Clinical_Data.loc[ds['Age at CT'] < 18.5, 'Age at CT'] = 0
Clinical_Data.loc[((ds['Age at CT']) >= 18.5) & ((ds['Age at CT']) < 24.9), 'Age at CT'] = 1
Clinical_Data.loc[((ds['Age at CT']) >= 24.9) & ((ds['Age at CT']) < 50), 'Age at CT'] = 2
Clinical_Data.loc[((ds['Age at CT']) >= 50) & ((ds['Age at CT']) < 60), 'Age at CT'] = 3
Clinical_Data.loc[((ds['Age at CT']) >= 60) & ((ds['Age at CT']) < 70), 'Age at CT'] = 4
Clinical_Data.loc[((ds['Age at CT']) >= 70) & ((ds['Age at CT']) < 80), 'Age at CT'] = 5
Clinical_Data.loc[ds['Age at CT'] > 80, 'Age at CT'] = 6

# Tobacco
Clinical_Data.loc[ds['Tobacco'] == 'No', 'Tobacco'] = 0
Clinical_Data.loc[ds['Tobacco'] == 'Yes', 'Tobacco'] = 1
Clinical_Data['Tobacco'] = Clinical_Data['Tobacco'].fillna('-') # -: Missing/not know

# Alcohol abuse	
Clinical_Data.loc[~Clinical_Data['Alcohol abuse'].isnull()] = 1 # 1: Has any kind of Alcohol abuse
Clinical_Data['Alcohol abuse'] = Clinical_Data['Alcohol abuse'].fillna(0) # 0: No Alcohol abuse (or Missing/not know)


# FRS 10-year risk (%)
Clinical_Data['FRS 10-year risk (%)'] = Clinical_Data['FRS 10-year risk (%)'].astype(str)
Clinical_Data['FRS 10-year risk (%)'] = Clinical_Data['FRS 10-year risk (%)'].str.replace("%", "")
Clinical_Data['FRS 10-year risk (%)'] = Clinical_Data['FRS 10-year risk (%)'].str.replace("<1", "0.5") # switch <1 to 0.5
Clinical_Data['FRS 10-year risk (%)'] = Clinical_Data['FRS 10-year risk (%)'].str.replace(">30", "5000") # >30 is 5000
Clinical_Data['FRS 10-year risk (%)'] = Clinical_Data['FRS 10-year risk (%)'].str.replace("X", "10000") # X is 10000
Clinical_Data['FRS 10-year risk (%)'] = Clinical_Data['FRS 10-year risk (%)'].astype(float)
Clinical_Data.loc[(Clinical_Data['FRS 10-year risk (%)']) < 5, 'FRS 10-year risk (%)'] = 0
Clinical_Data.loc[((Clinical_Data['FRS 10-year risk (%)']) >= 5) & ((Clinical_Data['FRS 10-year risk (%)']) < 10), 'FRS 10-year risk (%)'] = 1
Clinical_Data.loc[((Clinical_Data['FRS 10-year risk (%)']) >= 10) & ((Clinical_Data['FRS 10-year risk (%)']) < 15), 'FRS 10-year risk (%)'] = 2
Clinical_Data.loc[((Clinical_Data['FRS 10-year risk (%)']) >= 15) & ((Clinical_Data['FRS 10-year risk (%)']) < 20), 'FRS 10-year risk (%)'] = 3
Clinical_Data.loc[(Clinical_Data['FRS 10-year risk (%)']) >= 20, 'FRS 10-year risk (%)'] = 4
Clinical_Data.loc[(Clinical_Data['FRS 10-year risk (%)']) == 5000, 'FRS 10-year risk (%)'] = 5
Clinical_Data.loc[(Clinical_Data['FRS 10-year risk (%)']) == 10000, 'FRS 10-year risk (%)'] = 6

# FRAX 10y Fx Prob (Orange-w/ DXA)
Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)'] = pd.to_numeric(Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)'], errors='coerce') # force string to NaN
Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)'] = Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)'].fillna(10000)
Clinical_Data.loc[(Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) < 5, 'FRAX 10y Fx Prob (Orange-w/ DXA)'] = 0
Clinical_Data.loc[((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) >= 5) & ((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) < 10), 'FRAX 10y Fx Prob (Orange-w/ DXA)'] = 1
Clinical_Data.loc[((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) >= 10) & ((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) < 15), 'FRAX 10y Fx Prob (Orange-w/ DXA)'] = 2
Clinical_Data.loc[((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) >= 15) & ((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) < 20), 'FRAX 10y Fx Prob (Orange-w/ DXA)'] = 3
Clinical_Data.loc[((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) >= 20) & ((Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) < 25), 'FRAX 10y Fx Prob (Orange-w/ DXA)'] = 4
Clinical_Data.loc[(Clinical_Data['FRAX 10y Fx Prob (Orange-w/ DXA)']) > 25, 'FRAX 10y Fx Prob (Orange-w/ DXA)'] = 5


# FRAX 10y Hip Fx Prob (Orange-w/ DXA)
Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)'] = pd.to_numeric(Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)'], errors='coerce') # force string to NaN
Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)'] = Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)'].fillna(10000)
Clinical_Data.loc[(Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)']) < 5, 'FRAX 10y Hip Fx Prob (Orange-w/ DXA)'] = 0
Clinical_Data.loc[((Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)']) >= 2.5) & ((Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)']) < 5), 'FRAX 10y Hip Fx Prob (Orange-w/ DXA)'] = 1
Clinical_Data.loc[((Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)']) >= 5) & ((Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)']) < 10), 'FRAX 10y Hip Fx Prob (Orange-w/ DXA)'] = 2
Clinical_Data.loc[(Clinical_Data['FRAX 10y Hip Fx Prob (Orange-w/ DXA)']) >= 5, 'FRAX 10y Hip Fx Prob (Orange-w/ DXA)'] = 3

# Met Sx
Clinical_Data['Met Sx'] = Clinical_Data['Met Sx'].fillna('-') # -: Missing/not know
Clinical_Data.loc[ds['Met Sx'] == 'N', 'Met Sx'] = 0
Clinical_Data.loc[ds['Met Sx'] == 'Y', 'Met Sx'] = 1

Clinical_Data.head(500)
```
